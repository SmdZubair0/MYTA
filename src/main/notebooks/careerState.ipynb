````python
import requests
from bs4 import BeautifulSoup
from lxml import etree
````

````python
HEADERS = {
    "User-Agent": "Mozilla/5.0 (compatible; MYTA/1.0)"
}
````

````python
link = "https://www.linkedin.com/jobs/ai-junior-engineer-jobs-bengaluru"
````

````python
response= requests.get(link, headers=HEADERS)
````

````python
soup = BeautifulSoup(response.text).main
````

````python
soup
````

````python
soup.find_all(class_="ember-view")
````

````python
tree = etree.HTML(str(soup))
````

````python
soup.find("//li[contains(@class,\"ember-view\")]")
````

````python
all_jobs = tree.xpath("//a[contains(@class,\"base-card__full-link\")]")
all_jobs
````

````python
links = []
for job in all_jobs:
    links.append(job.get('href'))
````

````python
links
````

````python
response = requests.get(links[0], headers=HEADERS)
````

````python
e = etree.HTML(response.text)
````

````python
career_stage = None
job_type = None
for li in e.cssselect(".description__job-criteria-list li"):
    if "seniority level" in li.cssselect("h3")[0].text.lower():
        career_stage = li.cssselect("span")[0].text
    if "employment type" in li.cssselect("h3")[0].text.lower():
        job_type = li.cssselect("span")[0].text
````
